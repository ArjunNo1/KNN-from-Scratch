```python
##Implement knn from scratch
import numpy as np
def KNN_scratch(X,y,k,point):
    total_dist=[]
    for i in range(X.shape[0]):
        #calc dist of pt from all other pts
        dist=np.sqrt((X[i][0]-point[0])**2+(X[i][1]-point[1])**2)
        
        total_dist.append([dist,y[i]])
    total_dist.sort()
    
    total_dist=total_dist[:k]
    
    count_0,count_1=0,0
    for i in range(k):
        if total_dist[i][1]==0:
            count_0+=1
        else:
            count_1+=1
    print(count_0,count_1)
    if count_0>count_1:
        return 0
    else:
        return 1
```

```python
KNN_scratch(X,y,10,[-2,0])
```

```python
#import algo from sklearn
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=int(np.sqrt(X.shape[0])))
```

```python
classifier.fit(X,y)
```

```python
classifier.predict(np.array([-1,2]).reshape(1,-1))[0]
```
